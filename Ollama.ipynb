```python
from ollama import chat
```

```python
# First Prompt - create chatbot using ollama
response = chat(
    model="llama3.2:3b",
    messages=[
        {
            "role":"user",
            "content":"What is Artificial Intelligence?"
        }
    ]
)

print(response["message"]["content"])
```

```python
# Multiturn conversation
messages = [
    {
        "role":"user",
        "content":"What is Machine Learning?"
    }
]

response = chat(
    model="llama3.2:3b",
    messages=messages
)

answer = response["message"]["content"]

print(answer)

messages.append(
    {
        "role":"assistant",
        "content":answer
    }
)

messages.append(
    {
        "role":"user",
        "content":"Give one real-world example."
    }
)

response = chat(
    model="llama3.2:3b",
    messages=messages
)

print(response["message"]["content"])
```

```python
# System Prompt
response = chat(
    model="llama3.2:3b",
    messages=[
        {
            "role":"system",
            "content":"You are a computer science professor."
        },
        {
            "role":"user",
            "content":"Explain recursion."
        }
    ]
)

print(response["message"]["content"])
```

```python
# Zero-shot prompting
prompt = """
Classify the sentiment.

Sentence:
The lecture was excellent.
"""

response = chat(
    model="llama3.2:3b",
    messages=[
        {
            "role":"user",
            "content":prompt
        }
    ]
)

print(response["message"]["content"])
```

```python
# One-shot prompting
prompt = """
Example

Sentence:
The movie was fantastic.

Sentiment:
Positive

Sentence:
The assignment was confusing.

Sentiment:
"""

response = chat(
    model="llama3.2:3b",
    messages=[
        {
            "role":"user",
            "content":prompt
        }
    ]
)

print(response["message"]["content"])
```

```python
# Few shot prompting
prompt = """
Example 1

The food was delicious.

Positive

Example 2

The service was terrible.

Negative

Example 3

The class starts at 10 AM.

Neutral

Sentence

The exam was difficult but fair.

Sentiment
"""

response = chat(
    model="llama3.2:3b",
    messages=[
        {
            "role":"user",
            "content":prompt
        }
    ]
)

print(response["message"]["content"])
```

```python
# Temperature
from ollama import chat

response = chat(
    model="llama3.2:3b",
    messages=[
        {
            "role":"user",
            "content":"Suggest five startup ideas using AI."
        }
    ],
    options={
        "temperature":0.2
    }
)

print(response["message"]["content"])
```

```python
# Higher Temperature
response = chat(
    model="llama3.2:3b",
    messages=[
        {
            "role":"user",
            "content":"Suggest five startup ideas using AI."
        }
    ],
    options={
        "temperature":0.2
    }
)

print(response["message"]["content"])
```

```python
# Top p
response = chat(
    model="llama3.2:3b",
    messages=[
        {
            "role":"user",
            "content":"Write a short fantasy story."
        }
    ],
    options={
        "top_p":0.6
    }
)

print(response["message"]["content"])
```

```python
# Max tokens
response = chat(
    model="llama3.2:3b",
    messages=[
        {
            "role":"user",
            "content":"Explain Neural Networks."
        }
    ],
    options={
        "num_predict":40
    }
)

print(response["message"]["content"])
```

```python
# streaming output
from ollama import chat

stream = chat(
    model="llama3.2:3b",
    messages=[
        {
            "role":"user",
            "content":"Explain Generative AI."
        }
    ],
    stream=True
)

for chunk in stream:
    print(chunk["message"]["content"], end="", flush=True)
```

```python
# Simple chatbot
history = [
    {
        "role":"system",
        "content":"You are a helpful AI teaching assistant."
    }
]

while True:

    user = input("You: ")

    if user.lower() == "exit":
        break

    history.append(
        {
            "role":"user",
            "content":user
        }
    )

    response = chat(
        model="llama3.2:3b",
        messages=history
    )

    answer = response["message"]["content"]

    print("\nAssistant:", answer)

    history.append(
        {
            "role":"assistant",
            "content":answer
        }
    )
```
